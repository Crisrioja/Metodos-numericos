```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

"""Usamos sympy para obtener la serie truncada de Taylor de f (x) = ln(x) de cuarto orden con x0 = 1 y graficamos f(x) junto con el polinomio de Taylor"""

# Definimos la variable simbólica
x = sp.Symbol('x')

# Definimos la función f(x) = ln(x)
f = sp.ln(x)

# Punto de expansión x0 = 1 y orden de la serie de Taylor
x0 = 1
orden = 5  # Se usa 5 porque el término de orden n-1 es el último en la expansión

# Obtenemos la serie de Taylor de cuarto orden en x0 = 1
taylor_series = f.series(x, x0, orden).removeO()  # Se trunca en el orden 4

# Convertimos la serie de Taylor en una función numérica para graficar
taylor_func = sp.lambdify(x, taylor_series, 'numpy')
f_func = sp.lambdify(x, f, 'numpy')

# Crear valores de x para la gráfica (evitando x=0 por la singularidad de ln(x))
x_vals = np.linspace(0.2, 2, 400)
y_vals_f = f_func(x_vals)  # Valores de ln(x)
y_vals_taylor = taylor_func(x_vals)  # Valores de la aproximación de Taylor

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals_f, label=r"$\ln(x)$", color='blue')
plt.plot(x_vals, y_vals_taylor, label="Serie de Taylor (4to orden)", linestyle='--', color='red')

# Marcar el punto de expansión x0 = 1
plt.scatter([x0], [f_func(x0)], color='black', zorder=3, label=r'$x_0 = 1$')

# Configuración de la gráfica
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximación de Taylor de $f(x) = \ln(x)$ en $x_0 = 1$")
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()

# Mostrar la serie de Taylor obtenida en consola
print("Serie de Taylor de ln(x) en x0 = 1 (orden 4):")
sp.pprint(taylor_series)
```
