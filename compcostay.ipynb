<font color = red >Comparación de la función cos y la serie de taylor para apreciar su aproximación tal que

$f(x)=cos(x)$
$g(x)=1 - (x^2)/2 + (1/4!)x4 - (1/6!)x^6$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definimos la foncion que aproxima a la serie de taylor
def aproxtaylor(x):
    return 1 - (x**2)/2 + (x**4)/math.factorial(4) - (x**6)/math.factorial(6)

x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calcular los valores de cos(x) y la aproximación de Taylor
y_cos = np.cos(x)
y_approx = aproxtaylor(x)

# Graficar cos(x)
plt.plot(x, y_cos, label='cos(x)', color='blue')

# Graficar la aproximación de Taylor
plt.plot(x, y_approx, label='Aproximación de Taylor', linestyle='--', color='red')

# Configuracion de la gráfica
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de cos(x) y su aproximación de Taylor')
plt.legend()
plt.grid(True)

# Mostramos la gráfica
plt.show()
```
